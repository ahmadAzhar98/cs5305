# Using llm_utils Module

This notebook demonstrates how to import and use the `llm_utils` module
for initializing Azure OpenAI LLMs and embeddings.

## Step 1: Add Module to Path

First, add the `llm_utils` package location to your Python path.

````python
import sys
from pathlib import Path

# Add project root to path
module_path = str(Path("../").resolve())
if module_path not in sys.path:
    sys.path.insert(0, module_path)

print(f"Added to Python path: {module_path}")
````

## Step 2: Initialize Environment

Load environment variables and initialize the setup.

````python
from llm_utils import setup

setup()
````

## Step 3: Create Azure LLM

Create an Azure OpenAI language model instance.

````python
from llm_utils import create_azure_llm

# Create LLM with default settings
llm = create_azure_llm()

print("\nLLM created successfully!")
print(f"Type: {type(llm)}")
````

## Step 4: Use the LLM

Invoke the LLM with a prompt.

````python
# Simple prompt
response = llm.invoke("What is Python?")
print("Response:")
print(response.content)
````

## Step 5: Create Embeddings

Create an Azure OpenAI embedding model.

````python
from llm_utils import create_azure_embedding

# Create embeddings model
embeddings = create_azure_embedding()

# Embed a sample text
sample_text = "Machine learning is a subset of artificial intelligence"
vector = embeddings.embed_query(sample_text)

print(f"Text: {sample_text}")
print(f"Embedding dimension: {len(vector)}")
print(f"First 5 values: {vector[:5]}")
````

## Step 6: Use with LangChain Agents

Use the LLM with LangChain agents.

````python
from langchain.agents import create_agent
from langchain.tools import tool

@tool
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        result = eval(expression)
        return f"{expression} = {result}"
    except Exception as e:
        return f"Error: {e}"

# Create a simple agent using our LLM
agent = create_agent(
    model=llm,
    tools=[calculator],
    system_prompt="You are a helpful math assistant. Use the calculator tool when needed."
)

# Use the agent
from langchain.messages import HumanMessage
response = agent.invoke({
    "messages": [HumanMessage(content="What is 25 * 4?")]
})

print("Agent response:")
print(response["messages"][-1].content)
````

## Step 7: Use with Event Planner Module

Combine with the event_planner module for more complex workflows.

````python
from event_planner import create_coordinator_with_memory

# Create event planner coordinator with our LLM
coordinator = create_coordinator_with_memory(llm=llm)

print("Event planner coordinator created!")
print(f"Type: {type(coordinator)}")
````

## Configuration Options

You can customize the LLM creation with different parameters.

````python
# More deterministic responses (temperature = 0)
llm_deterministic = create_azure_llm(temperature=0.0)

# More creative responses (temperature = 1.0)
llm_creative = create_azure_llm(temperature=1.0)

# Different deployment type
# llm_audio = create_azure_llm(deployment_type="audio")

print("Created LLMs with different configurations:")
print(f"- Deterministic (temp=0.0): {llm_deterministic}")
print(f"- Creative (temp=1.0): {llm_creative}")
````
